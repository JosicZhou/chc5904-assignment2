```python
# import libraries
import os
import pandas as pd
import re # 正则表达式库，用于更准确地分割句子
```

```python
# open file

text_folder = '/content/drive/MyDrive/CHC_assignment2_ZhouJing_25096585g/text_data/'
project_folder = '/content/drive/MyDrive/CHC_assignment2_ZhouJing_25096585g/'

output_file = os.path.join(project_folder, 'rulin_data.csv')
```

```python
# store the city list and character list
city_list = ['南京', '揚州', '杭州', '蘇州', '濟南']
character_list = ['王冕', '范進', '周進', '荀玫', '杜少卿', '嚴監生', '嚴貢生', '馬二先生']

all_records = []

chapters_to_analyze = [str(i) for i in range(1, 21)]

#read all the data
for chapter_num in chapters_to_analyze:
    filename = f"ch{chapter_num.zfill(2)}.txt"
    file_path = os.path.join(text_folder, filename)

    if os.path.exists(file_path):
        print(f"Processing: {filename}")
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()

        sentences = re.split(r'([。？！])', text)

        sentences = [sentences[i] + sentences[i+1] for i in range(0, len(sentences)-1, 2)]

        # find city and character in every chapter
        for sentence in sentences:
            if not sentence.strip():
                continue

            for city in city_list:
                if city in sentence:

                    found_character = 'Not found'
                    for char in character_list:
                        if char in sentence:
                            found_character = char
                            break


                    record = {
                        'Chapter': chapter_num,
                        'City': city,
                        'Character': found_character,
                        'Activity_Summary': '',
                        'Original_Text': sentence.strip()
                    }
                    all_records.append(record)
    else:
        print(f"Cannot find {filename}")

# preview the data
df = pd.DataFrame(all_records)
print("data preview result")
print(df.head(10))

# save data
df.to_csv(output_file, index=False, encoding='utf-8-sig')

print("Done")
```
